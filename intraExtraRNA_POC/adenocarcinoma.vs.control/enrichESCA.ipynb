```python
ESCA_IntersectionDict = dict()
candidateGeneSet = set()
for key in degreeDict.keys():
    if 'ESCA' in key:
        ESCA_IntersectionDict[key] = degreeDict[key]
        candidateGeneSet.update( degreeDict[key] )
```
