```python
import pandas as pd
```

```python
%%time
countPathRoot = "/private/groups/kimlab/vikas/nanopore/promethion/barretts/analysis/complete-seq/R_results"
countHighPath = f'{countPathRoot}/normalized_counts_for_andy.csv'
highCountDF = pd.read_csv(countHighPath, index_col="rowname")
print(f'loading:\n{countHighPath}')
highCountDF.index = highCountDF.index.rename('geneId')
highCountNumSamples = highCountDF.shape[1] - 1
print(f'highCountDF.shape : {highCountDF.shape} highCountNumSamples: {highCountNumSamples}')

# set flag we need to filter colData
isHighCountDF = True

highCountDF.head()
```

```python
%%time
rootDir = "/private/groups/kimlab/aedavids/londonCalling2024/data"
colDataPath = f'{rootDir}/colData.high.adenocarcinoma.control.csv'
print(f'colDataPath {colDataPath}')
colDataDF = pd.read_csv(colDataPath)
print(f'colDataDF.shape : {colDataDF.shape}')
colDataDF.head(n=10)
```

# select the counts for high count adenocarcinoma or control samples

```python
selectCols = colDataDF.loc[:,'countType'] == "high"
highCountCols = colDataDF.loc[selectCols, 'sampleId']
highCountAdenocarcinomaControlDF = highCountDF.loc[:, highCountCols]
print(f'highCountAdenocarcinomaControlDF.shape : {highCountAdenocarcinomaControlDF.shape}' )
```

```python
assert colDataDF.shape[0] == highCountAdenocarcinomaControlDF.shape[1], f'ERROR dimensions do not match'
```

```python
path = f'{rootDir}/normalizedCounts.high.adenocarcinoma.control.csv'
highCountAdenocarcinomaControlDF.to_csv(path)
print(f'saved to\n{path}')
```
