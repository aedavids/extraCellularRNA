**bar chart over vie generated by chatGPT4** 
- x axis categorys Lung, LUAD, ...
- y ratio = (number of repeats) / (number that are not repeats)
- standard error of a proportion is ( p(1-p) / n) **0.5 ; AEDWIP this looks like the std
  * n = number of observations
  * p is proportion of the numerator of the ration. ie proportion of repeats
- error bar range = standard error * Z
  * Z = Z-score corresponding to your confidence level (e.g., 1.96 for 95% confidence)
 
**proportion stats**  
[6.3: The Sample Proportion](https://stats.libretexts.org/Bookshelves/Introductory_Statistics/Introductory_Statistics_(Shafer_and_Zhang)/06%3A_Sampling_Distributions/6.03%3A_The_Sample_Proportion#:~:text=at%20least%2030.-,The%20Sampling%20Distribution%20of%20the%20Sample%20Proportion,P%3D%E2%88%9Apqn.)

p = num of positive observations / total number of observations  
q =  1 - p  
n = number of observations  
std = sqrt( pq / n )

**error bar choices**  
1. standard deviation 
2. standard error (the standard deviation of the averages)
3. confidence interval  
    C.I = observed metric +/- the margin of error
    + factors affecting CI  
      a. number of observations  
      b. variablity  
      c. The confidence level: how much you can trust your confidence-interval calculation
    + [Khan Acadamey Confidence intervals and margin of error](https://www.khanacademy.org/math/ap-statistics/xfb5d8e68:inference-categorical-proportions/introduction-confidence-intervals/v/confidence-intervals-and-margin-of-error)

````python
def selectGenesWithCounts(df : pd.DataFrame) -> pd.DataFrame:
    '''
    make bar code plot testing easier. select rows that do not have zeros
    '''
    # create a logical data frame
    nonZeroCountsDF = df != 0
    print

    byRow = 1
    nonZeroCountsDF['rowSum'] = nonZeroCountsDF.sum(axis=byRow)

    numSampleCols = df.shape[1]
    print(f'numSampleCols : {numSampleCols}')
    selectGenesWithOutZerosRow = nonZeroCountsDF['rowSum'] == numSampleCols

    retDF = df.loc[selectGenesWithOutZerosRow, :]

    return retDF

def testSelectGenesWithCounts():
    retDF = selectGenesWithCounts( plotTestCountsDF.head() )
    display(retDF)

testSelectGenesWithCounts()
````

````python
def createBarChartPlotData(plotTestCountsDF) :
    nonZeroCountsDF = selectGenesWithCounts( plotTestCountsDF )
    nonZeroCountsDF = nonZeroCountsDF.head(n=6)
  
    typeDict = dict()
    typeDict['Lung'] = nonZeroCountsDF.loc[:, ['geneId'] + lungSampleIdsDF['sample_id'].to_list()]
    typeDict['LUAD'] = nonZeroCountsDF.loc[:, ['geneId'] + LUADSampleIdsDF['sample_id'].to_list()]
    
    #print( pp.pformat(typeDict) )
    
    # sum the transcript counts
    byRow = 1
    for key, df in typeDict.items():
        # all colunms except geneId
        sampleCols = ~ df.columns.isin( ['geneId'] )
        df['rowSum'] = df.loc[:, sampleCols].sum(axis=byRow)
    
    # print('\n******** sum rows')
    # print( pp.pformat(typeDict) )
    
    # assume the first 3 cols are repeats
    for key, df in typeDict.items():
        # all colunms except geneId
        sampleCols = ~ df.columns.isin( ['geneId'] )
        n = int(df.shape[0] / 2)
        isRepeat = [True]*n + [False]*n
        df['isRepeat'] = isRepeat
    
    # print('\n******** sum rows')
    # print( pp.pformat(typeDict) )

    return typeDict

barChartDict = createBarChartPlotData(plotTestCountsDF)
print( pp.pformat(barChartDict) )
````
