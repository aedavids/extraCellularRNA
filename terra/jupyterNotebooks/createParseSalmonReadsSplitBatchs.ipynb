# createParseSalmonReadsSplitBatchs
create a set of csv batchs sets for extraCellularRNA/terra/deseq/bin/parseSalmonReadsSplitCLI.sh

```python
import pandas as pd
import pathlib as pl
```

```python
dataDir = "../deseq/bin/data"
csvFile = dataDir + "/sparkGTEx-trainQuantFiles.csv"
```

```python
trainDF = pd.read_csv( csvFile )
trainDF.head()
```

```python
path="gs://anvil-gtex-v8-hg38-edu-ucsc-kim-lab-spark/quant/sparkGTEx-train.out/parseSalmonReadsSelectCountsCLI.out/"
trainDF['splitFile'] = path + trainDF['sampleName']
pd.set_option('max_colwidth', None) # show full width of showing cols
trainDF = trainDF.loc[:,['sampleName', 'splitFile']]
trainDF = trainDF.rename(columns = {'sampleName':'sampleName', 'splitFile':'source'})
trainDF.head()
```

```python
## create batch files
```

```python
numRows = trainDF.shape[0]
numBatches = 50
numRows = numRows // numBatches
numRows = int(numRows)
remainingRows = numRows % numBatches
print( "numRows:{} numBatches:{} numRows:{} remainingRows:{}"
      .format(numRows, numBatches, numRows, remainingRows))
```

```python
outputDir = pl.Path( dataDir + "/quant/parseSalmonReadsSplitCLI/quantFilesCSV")
outputDir.mkdir(exist_ok=True, parents=True)

i = 0
while i < numBatches:
    start = i * numRows
    end = start + numRows
    print("\ni:{} start:{} end:{}".format(i, start,end))
    df = trainDF.iloc[ start:end ]
    f = "sparkGTEx-trainQuantFiles.csv.batch-{}".format(i)
    path = outputDir.joinpath(f)
    print("write to:{}".format(path))
    df.to_csv(path, index=False)

    # Increment the batch number
    i += 1
```
