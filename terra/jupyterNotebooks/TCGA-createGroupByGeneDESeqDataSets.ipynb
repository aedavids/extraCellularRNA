# create the group by gene id training sets

````python
groupByFiles = ! ls "$localGroupByTmp"
assert len(groupByFiles) == 33, "ERROR missing data files"
groupByFiles
````

````python
cohortDFDict = {}
for f in groupByFiles:
    # f = TCGA_UVM_ControlledAccess_V1-0_DATA_edu_ucsc_kim_lab.NumReads.csv'
    cohort = f.split("_")[1]
    p = localGroupByTmp + "/" + f
    print("cohort:{} p: {}".format(cohort, p))
    df = pd.read_csv(p)
    cohortDFDict[cohort] = df
````

## Explore data
This should make it easier to understand how createSplit() works

````python
accGroupByDF = cohortDFDict['ACC']
display(accGroupByDF.iloc[:, 0:4].head())
testColDataDF = colDFDict['TCGA-TestColData.csv']
testColDataDF.loc[:, ['sample_id', 'Cohort']].head()
````

````python
def createSplit(colDataDF, cohortDFDict):
    #DEBUG = 0
    retDF = pd.DataFrame()
    for cohort in cohortDFDict.keys():
#         if DEBUG > 1:
#             break
            
        groupByCohortDF = cohortDFDict[cohort]
        
        selectCohortRows = colDataDF['Cohort'] == cohort
        sampleIdDF = colDataDF.loc[selectCohortRows, ['sample_id']]
        #print(sampleIdDF.shape)
        print("cohort:{} shape:{}".format(cohort, sampleIdDF.shape))
        #display(sampleIdDF)
        groupBySamplesDF = groupByCohortDF.loc[:, sampleIdDF['sample_id']]
        
        if retDF.empty:
            #print("init")
            retDF = groupBySamplesDF
        else:
            byColumns = 1
            tmpDF = pd.concat([retDF, groupBySamplesDF], axis=byColumns)
            retDF = tmpDF
            
#         DEBUG += 1

    # DESeq requires col idx order of count data frame matches the row idx of colData
    tmpDF = retDF.loc[:, colDataDF['sample_id']]
    retDF = tmpDF
    
    return retDF
    
    
testColDataDF = colDFDict['TCGA-TestColData.csv']
groupByTestDF = createSplit(testColDataDF, cohortDFDict)
print("groupByTestDF.shape:{}".format(groupByTestDF.shape))
display(groupByTestDF.head())
#print(groupByTestDF.columns)

# make sure there is no missing count data
assert groupByTestDF.isnull().sum().sum() == 0, "WARN: looks like there is missing data. Is there an concat() index problem"

# make sure the dimensions are correct
numGroupByCols = groupByTestDF.shape[1]
numColDataRows = testColDataDF.shape[0]
assert numGroupByCols  == numGroupByCols, "ERROR numGroupByCols:{} != numColDataRows:{}".format(numGroupByCols,numColDataRows)

# DESeq requires the count data columns to be in the same order as the the colData rows
groupByTestColumns = groupByTestDF.columns
colDataTestRows = testColDataDF['sample_id']
assert (groupByTestColumns == colDataTestRows).all(), "ERROR the columns are in the wrong order"
````
