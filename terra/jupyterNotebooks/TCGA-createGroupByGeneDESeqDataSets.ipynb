## implementation 

````python
def validateGeneIds(refGenesDF, groupBySamplesDF):
    '''
    all cohorts should have the same number of rows, the gene ids should be in the same order
    '''
    
    rows = 0
    assert refGenesDF.shape[rows] == groupBySamplesDF.shape[rows], "refGenes and groupBySampleDF have diff num rows"
    errorMsg = "groupBySamples do not match ref"
    assert (refGenesDF.loc[:, ["geneId"]].equals(groupBySamplesDF.loc[:, ["geneId"]]) ) , errorMsg
    
def createSplit(colDataDF, cohortDFDict):
    '''
    colDataDF contains list of samples to select
    '''
    print("colDataDF.shape: {}".format(colDataDF.shape))
    retDF = pd.DataFrame()
    for cohort in cohortDFDict.keys():
        groupByCohortDF = cohortDFDict[cohort]
        
        selectCohortRows = colDataDF['Cohort'] == cohort
        sampleIdDF = colDataDF.loc[selectCohortRows, ['sample_id']]
        print("cohort:{} shape:{}".format(cohort, sampleIdDF.shape))
 
        groupBySamplesDF = groupByCohortDF.loc[:, sampleIdDF['sample_id'] ]

        if retDF.empty:
            retDF = groupBySamplesDF
            # all the Cohort should be identidical we can choose which ever like as the ref
            RefGenesDF = groupByCohortDF.loc[:, ['geneId']]
        else:
            validateGeneIds(RefGenesDF, groupByCohortDF)
            byColumns = 1
            tmpDF = pd.concat([retDF, groupBySamplesDF], axis=byColumns)
            retDF = tmpDF

    # DESeq requires col idx order of count data frame matches the row idx of colData
    # we also want prepend the geneId column
    tmpDF = retDF.loc[:, colDataDF['sample_id']]
    retDF = tmpDF
    tmpDF = pd.concat([RefGenesDF, retDF], axis=byColumns)
    retDF = tmpDF
    print("return retDF.shape:{}".format(retDF.shape))
    
    return retDF
````

````python
def validate(groupByDF, colDataDF) :
    print("groupByDF.shape:{}".format(groupByDF.shape))
    display(groupByDF.head())
    #print(groupByDF.columns)

    # make sure there is no missing count data
    assert groupByDF.isnull().sum().sum() == 0, "WARN: looks like there is missing data. Is there an concat() index problem"

    # make sure the dimensions are correct
    # we have an extra 'geneId' col
    numGroupByCols = groupByDF.shape[1] - 1
    numColDataRows = colDataDF.shape[0]
    assert numGroupByCols  == numColDataRows, "ERROR numGroupByCols:{} != numColDataRows:{}".format(numGroupByCols,numColDataRows)

    # DESeq requires the count data columns to be in the same order as the the colData rows
    groupByColumnList = groupByDF.columns.to_list()
    groupByColumnList.remove('geneId')
    colDataRowLists = colDataDF['sample_id'].to_list()
    assert groupByColumnList == groupByColumnList, "ERROR the columns are in the wrong order"   
    
#validate(groupByTestDF, testColDataDF)
````

## run all

````python
%%time
def runAll(colDFDict, cohortDFDict):
    retDict = {}
    for colDFKey in colDFDict.keys():
        #'TCGA-TestColData.csv'
        dataSetName = colDFKey.split("-")[1].split(".")[0]
        dataSetName = dataSetName[0: -len('colData')] # remove colData
        print("\ncolDFKey: {} dataSetName: {}".format(colDFKey, dataSetName))
        
        colDataDF = colDFDict[colDFKey]
        groupByDF = createSplit(colDataDF, cohortDFDict)
        validate(groupByDF, colDataDF)

        retDict[dataSetName] = groupByDF
        
    return retDict

groupByGeneDataSetsDict =  runAll(colDFDict, cohortDFDict)
groupByGeneDataSetsDict.keys()
````
