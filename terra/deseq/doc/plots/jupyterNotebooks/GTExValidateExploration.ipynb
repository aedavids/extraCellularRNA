# get Top 25

```python
lungResults = dataRoot + "/" + "Validate_Lung_vs_all_results.csv"
lungDF = pd.read_csv(lungResults, skiprows=7)
lungDF.head()
```

```python
padjThreshold = 0.01
selectSignifigantRowsPS = lungDF.loc[:,"padj"] < padjThreshold
print("number of statistically signifigant genes: {}".format(selectSignifigantRowsPS.sum()))

lungLFCSignatureGenesDF = lungDF.loc[ selectSignifigantRowsPS,: ]\
    .sort_values("log2FoldChange", ascending=False)
    
n = 25
topLungLFCSignatureGenesDF = lungLFCSignatureGenesDF.head( n )
topLungLFCSignatureGenesDF
```

```python
lfcThreshold = 2.0
selectLFCPS = lungLFCSignatureGenesDF.loc[:,"log2FoldChange"] >= lfcThreshold
print("number of genes with (lfc >= {} and padj < {}) : {}"\
      .format(lfcThreshold, padjThreshold, selectLFCPS.sum()))
print("len(selectSignifigantRowsPS):{}".format(len(selectSignifigantRowsPS)))
print("len(selectLFCPS):{}".format(len(selectLFCPS)))


lungBaseMeanSignatureGenesDF = lungLFCSignatureGenesDF.loc[ selectLFCPS,: ]\
    .sort_values("baseMean", ascending=False)
    
topLungBaseMeanSignatureGenesDF = lungBaseMeanSignatureGenesDF.head( n )
topLungBaseMeanSignatureGenesDF
```

```python
def findSignatureGenes(deseqDF, topN=25, lfcThreshold=2.0, padjThreshold=0.01):
    # find genes that are statistically signifigant
    selectSignifigantRowsPS = deseqDF.loc[:,"padj"] < padjThreshold
    print("number of statistically signifigant genes: {}".format(selectSignifigantRowsPS.sum()))

    deseqLFCSignatureGenesDF = deseqDF.loc[ selectSignifigantRowsPS,: ]\
                                        .sort_values("log2FoldChange", ascending=False)
    
    # find the genes that are over expresed 
    selectLFCPS = deseqLFCSignatureGenesDF.loc[:,"log2FoldChange"] >= lfcThreshold
    deseqBaseMeanSignatureGenesDF = lungLFCSignatureGenesDF.loc[ selectLFCPS,: ]\
                            .sort_values("baseMean", ascending=False)
    
    topSignatureGenesDF = deseqBaseMeanSignatureGenesDF.head( topN )
    return topSignatureGenesDF
```

```python
lungResults = dataRoot + "/" + "Validate_Lung_vs_all_results.csv"
lungDF = pd.read_csv(lungResults, skiprows=7)
topLungSignatureGenesDF = findSignatureGenes(lungDF)
topLungSignatureGenesDF
```
