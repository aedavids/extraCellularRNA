## Genes that are signifigant and have high LFC may not be good signatue genes
We want our discriminator to have the widest possible margin between different classes. Notice some the the genes with high LFC and small p-adj have very small baseMean

```python
padjThreshold = 0.01
selectSignifigantRowsPS = lungDF.loc[:,"padj"] < padjThreshold
print("number of statistically signifigant genes: {}".format(selectSignifigantRowsPS.sum()))

lungLFCSignatureGenesDF = lungDF.loc[ selectSignifigantRowsPS,: ]\
    .sort_values("log2FoldChange", ascending=False)
    
n = 25
topLungLFCSignatureGenesDF = lungLFCSignatureGenesDF.head( n )
topLungLFCSignatureGenesDF
```

```python
def findSignatureGenes(deseqDF, topN=25, lfcThreshold=2.0, padjThreshold=0.01):
    # find genes that are statistically signifigant
    selectSignifigantRowsPS = deseqDF.loc[:,"padj"] < padjThreshold
    print("number of statistically signifigant genes: {}".format(selectSignifigantRowsPS.sum()))

    deseqLFCSignatureGenesDF = deseqDF.loc[ selectSignifigantRowsPS,: ]\
                                        .sort_values("log2FoldChange", ascending=False)
    
    # find the genes that are over expresed 
    selectLFCPS = deseqLFCSignatureGenesDF.loc[:,"log2FoldChange"] >= lfcThreshold
    print("number of genes with log2FoldChange >= {} = {}"\
            .format( lfcThreshold, selectLFCPS.sum() ))
    
    deseqBaseMeanSignatureGenesDF = deseqLFCSignatureGenesDF.loc[ selectLFCPS,: ]\
                            .sort_values("baseMean", ascending=False)
    
    topSignatureGenesDF = deseqBaseMeanSignatureGenesDF.head( topN )
    return topSignatureGenesDF
```

```python
def loadFindSaveSignatureGenes(tissueId):
    resultsFile = dataRoot + "/" + "Validate_" + tissueId + "_vs_all_results.csv"
    resultsDF = pd.read_csv(resultsFile, skiprows=7)
    topResulsSignatureGenesDF = findSignatureGenes(resultsDF)

    resultsOutput = dataRoot + "/" + "signatureGenesValidate" + tissueId + ".csv"
    topResulsSignatureGenesDF.to_csv(resultsOutput, index=False)
    print("saved to file: {}".format(resultsOutput))

    return topResulsSignatureGenesDF    
```
