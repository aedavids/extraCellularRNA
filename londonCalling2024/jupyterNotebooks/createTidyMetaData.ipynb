```python
# porting hack
# last night we only cleaned up the HighCount data
# countDF = countHighCountDF
# numSamples = highCountNumSamples
countDF = countAllCountDF
numSamples = allCountNumSamples
```

```python
# load colData.csv
# drop incomplete rows. They where used to create whitespace in the orginal spread sheet
colDataPath = f'{rootDir}/colDataIncomplete.csv'
print(f'colDataPath {colDataPath}')
rawColDataDF = pd.read_csv(colDataPath)
print(f'rawColDataDF.shape : {rawColDataDF.shape}')
rawColDataDF.head(n=10)
```

```python
# drop rows that where originally whitespace in spread sheet
# sample = NaN
# select rows where sampleId is not Nan
selectRows = ~rawColDataDF.loc[:,'sample'].isna()
print(f'sum {sum(selectRows)}')
colDataDF = rawColDataDF.loc[selectRows, :]
print(f'colDataDF.shape : {colDataDF.shape}')
display(colDataDF)
```

```python
# fix data types
colDataDF.loc[:, 'batchID'] = colDataDF['batchID'].astype(str)
colDataDF.loc[:, 'sample'] = colDataDF['sample'].astype(str)

# pd.options.mode.copy_on_write = True
# colDataDF.loc[:, 'sex'] = colDataDF['sex'].astype('category')
# colDataDF.loc[:, 'aedwip_race'] = colDataDF['race'].astype('category')
# colDataDF.loc[:,'aedwip_histology'] = colDataDF.loc[:, 'histology'].astype('category')

print(colDataDF.info())

display(colDataDF)
```

# add a sampleId that matches the count data colNames

```python
pd.set_option('display.max_colwidth', None)  
colDataDF.loc[:, 'suffix'] = colDataDF.loc[:, 'sample'].str[-5:]
colDataDF.loc[:, ['sample', 'suffix']]
```

```python
sampleIdDF = pd.DataFrame( {'sampleId' : countDF.columns[1:] } )
sampleIdDF
```

```python
#aedwipDF.loc[:, 'sampleId'].str.split(".")
```

```python
# regx 
# _: Matches the literal underscore character.
# ([^\.]+): Captures one or more characters that are not a dot. This is enclosed in 
# parentheses to create a capturing group, which str.extract will use to extract the value.
sampleIdDF['sampleIdToken'] = sampleIdDF.loc[:, 'sampleId'].str.extract(r'_([^\.]+)')
sampleIdDF
```

```python
sampleIdDF['suffix'] = sampleIdDF.loc[:, 'sampleIdToken'].str[-5:]
sampleIdDF
```

```python
saveDF = pd.merge(colDataDF, sampleIdDF, on='suffix', how='inner')
saveDF
```

# add a column 'countType'
either high or low. 

```python
countHighCountDF.columns
```

```python
numRows = saveDF.shape[0] 
# countAllDF has more samples than countHighCountDF
countTypes = ["low"]*numRows
print(type(countTypes))
print(len(saveDF))
print(f'len(countTypes) : {len(countTypes)} saveDF.shape : {saveDF.shape}')
saveDF["countType"] = countTypes

selectHighRows =  saveDF.loc[:, "sampleId"].isin( countHighCountDF.columns )
saveDF.loc[selectHighRows, "countType"] = "high"

saveDF
```

# reorder the columns

```python
newOrder = ['sampleId', 'countType', 'sex', 'age', 'race', 'histology', 'batchID', 
            'sample', 'suffix','sampleIdToken', 'barcode']
saveDF = saveDF.loc[:, newOrder]
saveDF.head()
```
