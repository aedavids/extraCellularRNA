# expected set format
- ID (first column) 
    The first column is the ID column, the unique name of the set. The format is '<span style="color:red">membership:</span>your-entity-name<span style="color:red">_set_id</span>'. The parts in red are required exactly as typed. The entity-name is whatever the entity you are grouping together. For the Data-QuickStart it will be specimens, but if your workflow processes samples, it could be samples. 


- Entity column
    The second column is the entity you're grouping into sets. The header must match the first column header of the table your workflow will take its single inputs from.

Given a sample.tsv file  our sample_set file is 
```
(base) $ cat sample_set_membership.tsv 
membership:sample_set_id	sample
set-d-4-panc	GTEX-11GSP-0426-SM-5A5KX
set-d-4-panc	GTEX-11I78-0626-SM-5A5LZ
set-d-4-panc	GTEX-11LCK-0226-SM-5A5M6
set-d-4-panc	GTEX-11NSD-0526-SM-5A5LT
```

# Create Tissue specific  sets
setDF is in the correct format. We just need to append it to any previously sets define sets. These other sets could have been defined using the terra web form or a notebook running on terra

````python
# selectRows = sampleDF.loc[:, "tissue_id"] == tissueTypes[0]
#     tDF = sampleDF.loc[selectRows, :]

setDF = sampleDF.loc[ :, ['tissue_id',"entity:sample_id"] ].sort_values( by=['tissue_id'] )
setDF.head()                                                                     
````

````python
setId = "membership:" +  entityName  + "_set_id"
terraColNames = [ setId , entityName ]
setDF.columns = terraColNames
print("setDF.shape:{}".format(setDF.shape))
setDF.head()
````

````python
setDF.tail()
````

# example of how we might create a special set
This is really redundent. it is already in setDF

````python
pancTissue = 'Pancreas'
pancRows = setDF.loc[:, setId] == pancTissue
pancSetDF = setDF.loc[ pancRows ]
print(pancSetDF.count())
pancSetDF.head()
````

# Ex. 2 creating a special data set
This is more practicle. we want to define a small set of panc sample we can use to test or workflow with

````python
n = 10
pancSet10DF = pancSetDF.sample(n).copy(deep=True)
pancSet10DF[setId] = pancTissue + "_" + str(n)
pancSet10DF
````

````python
 #pancSet10DF.loc[:, "sample"]
pancSet10DF["sample"].to_list()
````

````python
print( pancSet10DF["sample"] )
print()
selectRows = pancSetDF.loc[:, "sample"].isin( pancSet10DF["sample"] )
pancSetDF.loc[selectRows, :]
````

````python
assert 10 == sum( pancSetDF.loc[selectRows, "membership:sample_set_id"] == pancTissue )
````

## Save the the two data sets as a TSV file
<span style="color:red">Append, sets could have been defined using web form</span>

````python
setMembershipPath = rootDir +  "/" +  entityName + "_set_membership.tsv" 
print(setMembershipPath)
origSetMembshipDF = pd.read_csv(setMembershipPath, delimiter='\t')
print( "origSetMembshipDF.shape:{}".format(origSetMembshipDF.shape))
# pd.set_option("max_colwidth", 40)
# with pd.option_context("display.max_rows", 10, "display.max_columns", None):
with pd.option_context("max_colwidth", None):
    print(origSetMembshipDF.head())
    print(origSetMembshipDF.tail())
````

````python
# we already defined the panc data set before. make sure we do not define it a second time
print("setDF.shape:{}".format(setDF.shape))
pancTissue = 'Pancreas'
notPancRows = setDF.loc[:, setId] != pancTissue
tissueSets =  setDF.loc[notPancRows, :]
print("tissueSets.shape:{}".format(tissueSets.shape))
assert tissueSets.shape[0] == 17054
````

````python
finalDataSetDF = pd.concat( [origSetMembshipDF, tissueSets, ] )
print("origSetMembshipDF.shape:{}".format(origSetMembshipDF.shape))
print("finalDataSetDF.shape:{}".format(finalDataSetDF.shape))
````
