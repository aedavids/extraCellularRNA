```python
import pandas as pd 
# display all columns and rows
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)

from analysis.hyperParameterTunningNotebookUtilities import root, outDir, notebookName, evaluate
```

```python
def evaluateBestCuratedDegree1(resultsDir : list[str], metric:str='sensitivity', threshold:float=0.7):
    
    stageName = "best10CuratedDegree1"
    bestCuratedDegree1Metrics1DF, bestCuratedDegree1BellowThresholdDF = evaluate(
                                                        root=root,
                                                        outDir=outDir,
                                                        resultsDirs=resultsDir, 
                                                        outFilePrefix=stageName,
                                                        metric=metric,
                                                        stageName=stageName,
                                                        threshold=threshold,
                                                        verbose=False,
                                                        )
best10CuratedDegree1_ce467ff_degree1LUSCDict = {
   
    "baseReference" : [ 
        # "best10CuratedDegree1",
        # "best10CuratedDegree1_6457b56", # bug we did not sort in decending order
        # "best10CuratedDegree1.noArtifical", # does not artifically enrich        
        "best10CuratedDegree1_ce467ff"],   

    "Curated LUSC" : [
        "best10CuratedDegree1_ce467ff.degree1LUSC1",  
        "best10CuratedDegree1_ce467ff.degree1LUSC2",  
        "best10CuratedDegree1_ce467ff.degree1LUSC3",        
        "best10CuratedDegree1_ce467ff.degree1LUSC5",
        "best10CuratedDegree1_ce467ff.degree1LUSC10",
        ],
 }

for name,resultsDir in best10CuratedDegree1_ce467ff_degree1LUSCDict.items() :
    print(f'\n###### {name}')
    evaluateBestCuratedDegree1(resultsDir, metric='sensitivity', threshold=0.7)
```

## False Negatives

```python
for name,resultsDir in best10CuratedDegree1_ce467ff_degree1LUSCDict.items() :
    print(f'\n###### {name}')
    # set threshold to display all false negatives
    evaluateBestCuratedDegree1(resultsDir, metric='fn', threshold=15000)
```
